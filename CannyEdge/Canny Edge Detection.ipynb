### First we make a Gaussian Filter and apply it on the original image

```python
def make_gaussian_filter(s, sigma):
    max_x = max_y = s // 2
    min_x = min_y = -max_x

    filter = np.zeros((s, s))

    for x in range(min_x, max_x + 1):
        for y in range(min_x, max_y + 1):
            v = (1 / (2 * np.pi * (sigma ** 2))) * np.exp(-(x ** 2 + y ** 2) / (2 * (sigma ** 2)))
            filter[x - min_x, y - min_y] = v

    return filter
```

```python
gaussian_filter = make_gaussian_filter(3, 0.5)
```

```python
np.sum(gaussian_filter)
```

```python
img_filtered = cv2.filter2D(gray_img, -1, gaussian_filter)
```

```python
plt.figure()
plt.subplot(121)
plt.imshow(gray_img, cmap = 'gray')

plt.subplot(122)
plt.imshow(img_filtered, cmap = 'gray')

plt.show()
```

## Derivation of Gaussian
---
#### With some modification we can achieve application of guassain filter and intensity at either X-axis or Y-axis together

```python
def get_x_gaussian(s, sigma):
    max_x = max_y = s // 2
    min_x = min_y = -max_x

    x_gaussian = np.zeros((s, s))

    for x in range(min_x, max_x + 1):
        for y in range(min_x, max_y + 1):
            v = (-x / (2 * np.pi * (sigma ** 4))) * np.exp(-( x ** 2 + y ** 2)/(2 * (sigma ** 2)))
            x_gaussian[x - min_x, y - min_y] = v

    return x_gaussian
```

```python
def get_y_gaussian(s, sigma):
    max_x = max_y = s // 2
    min_x = min_y = -max_x

    y_gaussian = np.zeros((s, s))

    for x in range(min_x, max_x + 1):
        for y in range(min_x, max_y + 1):
            v = (-y / (2 * np.pi * (sigma ** 4))) * np.exp(-( x ** 2 + y ** 2)/(2 * (sigma ** 2)))
            y_gaussian[x - min_x, y - min_y] = v

    return y_gaussian
```

```python
x_gaussian = get_x_gaussian(3, 0.5)
y_gaussian = get_y_gaussian(3, 0.5)
```

```python
x_gaussian_img = cv2.filter2D(gray_img, -1, x_gaussian)
y_gaussian_img = cv2.filter2D(gray_img, -1, y_gaussian)
```

```python
np.array_equal(x_gaussian_img, y_gaussian_img)
```

```python
plt.figure()
plt.subplot(131)
plt.imshow(img)

plt.subplot(132)
plt.imshow(x_gaussian_img, cmap = 'gray')

plt.subplot(133)
plt.imshow(y_gaussian_img, cmap = 'gray')
```

```python
plt.imsave('x_gaussian_img.jpg', x_gaussian_img, cmap = 'gray')
plt.imsave('y_gaussian_img.jpg', y_gaussian_img, cmap = 'gray')
```

## Intensity Gradient & Edge Direction

```python
GM = (x_gaussian_img ** 2 + y_gaussian_img ** 2) ** 0.5
```

```python
plt.imsave('Gradient_Maxima.jpg', GM, cmap = 'gray')
```

```python
ED = np.rad2deg(np.arctan2(Iy, Ix))
```

```python
plt.imsave('EdgeDirection.jpg', ED, cmap = 'gray')
```

```python
plt.figure()
plt.subplot(121)
plt.imshow(GM, cmap = 'gray')

plt.subplot(122)
plt.imshow(ED, cmap = 'gray')

plt.show()
```

```python
def NMS(Gm,Gd):
    Gd_bin = 45*(np.round(Gd/45))
    Gm_NMS = np.zeros(Gm.shape)
    numRows,numCols = Gm.shape[0],Gm.shape[1]
    a,b=0.,0.
    for r in range(1,numRows-1):
        for c in range(1,numCols-1):
            ang = Gd_bin[r,c]
            if ang == 180. or ang == -180. or ang == 0.0 or ang == -0.0:
                a,b = Gm[r-1,c], Gm[r+1,c]
            elif ang == 45. or ang == -135.:
                a,b = Gm[r+1,c+1],Gm[r-1,c-1]
            elif ang == 90. or ang == -90.:
                a,b = Gm[r,c-1],Gm[r,c+1]
            elif ang == 135. or ang == -45.:
                a,b = Gm[r-1,c+1],Gm[r+1,c-1]
            else:
                print('Error')
                return
            if Gm[r,c] >= a and Gm[r,c] >= b:
                Gm_NMS[r,c] = Gm[r,c]
    return Gm_NMS
```

```python
NMS_image = NMS(GM, ED)
```

```python
plt.figure()
plt.subplot(121)
plt.imshow(GM, cmap = 'gray')

plt.subplot(122)
plt.imshow(NMS_image, cmap = 'gray')

plt.show()
```
