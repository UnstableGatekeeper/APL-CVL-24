```python
cv2.imshow('Albert Einstein', img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
print(cv2.__file__)
```

```python
tulips_img = cv2.imread('tulips.jpg')
```

```python
cv2.imshow('A bunch of tulips', tulips_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
plt.imshow(tulips_img)
```

```python
reverse_tulips_img = tulips_img[:, :, ::-1]
```

```python
cv2.imshow('A bunch of tulips', reverse_tulips_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
plt.imshow(reverse_tulips_img)
```

```python
original_tulips = cv2.imread('tulips.jpg', cv2.COLOR_BGR2RGB)
```

```python
cv2.imshow('Converted reading', original_tulips)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
