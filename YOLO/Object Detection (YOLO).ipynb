### Libraries we'll be using

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
import cvlib as cv # pip install cvlib
from cvlib.object_detection import draw_bbox
```

## Face Detection

```python
people_image = cv2.imread(r'People.jpg')
```

```python
plt.imshow(people_image[:,:,::-1])
```

```python
bbox, conf = cv.detect_face(people_image)
```

```python
bbox
```

```python
conf
```

```python
labels = []
for i in range (len(conf)):
    labels.append('person')
```

```python
people_image_foo = draw_bbox(people_image, bbox, labels, conf)
```

```python
plt.imshow(people_image_foo[:,:,::-1])
```

## People Detection
---

```python
standing_image = cv2.imread(r'People_standing.jpg')
```

```python
plt.imshow(standing_image[:,:,::-1])
```

```python
bbox, labels, conf = cv.detect_common_objects(standing_image)
```

```python
standing_image_foo = draw_bbox(standing_image, bbox, labels, conf)
```

```python
plt.imshow(standing_image_foo[:,:,::-1])
```
