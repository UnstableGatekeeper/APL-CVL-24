```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal
```

```python
original_img = cv2.imread('bird.jpg')
```

```python
original_img.shape
```

```python
plt.imshow(original_img)
```

```python
img = cv2.imread('bird.jpg', cv2.IMREAD_GRAYSCALE)
```

```python
img.shape
```

```python
plt.imshow(img, cmap = 'gray')
```

```python
filter = np.ones((11, 11))/255
filter
```

```python
blur = signal.convolve2d(img, filter, boundary = 'symm', mode = 'same')
```

```python
plt.figure()
plt.subplot(121)
plt.imshow(img,cmap='gray')

plt.subplot(122)
plt.imshow(blur,cmap='gray')

plt.show()
```

```python
x_mask = np.array([[-1,0,1],
                   [-1,0,1],
                   [-1,0,1]])
```

```python
y_mask = np.array([[-1,-1,-1],
                   [0,0,0],
                   [1,1,1]])
```

```python
x_mask_img = signal.convolve2d(blur, x_mask, boundary = 'symm', mode = 'same')
y_mask_img = signal.convolve2d(blur, y_mask, boundary = 'symm', mode = 'same')
```

```python
plt.figure()
plt.subplot(131)
plt.imshow(img,cmap='gray')

plt.subplot(132)
plt.imshow(x_mask_img,cmap='gray')

plt.subplot(133)
plt.imshow(y_mask_img,cmap='gray')
plt.show()
```

```python
GM = (x_mask_img**2 + y_mask_img**2)**0.5
```

```python
plt.figure()
plt.subplot(121)
plt.imshow(img,cmap='gray')

plt.subplot(122)
plt.imshow(GM>10,cmap='gray')

plt.show()
```
