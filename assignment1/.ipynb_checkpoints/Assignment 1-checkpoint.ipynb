```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

#### Original image

```python
img = cv2.imread('albert-einstein_gray.jpg', cv2.IMREAD_GRAYSCALE)
img.shape
```

```python
plt.imshow(img)
```

#### Resizing a image

```python
resized_img = cv2.resize(src= img, fx = 0.5, fy = 2, dsize = None)
plt.imshow(resized_img)
```

```python
resized_img.shape
```

#### Rotation

```python
height, width = img.shape
print(f'Height = {height}\n Width = {width}')
```

```python
img_center = (height // 2, width // 2)
img_center
```

```python
rotation_matrix = cv2.getRotationMatrix2D(img_center, 100, 0.5)
```

```python
rotated_img = cv2.warpAffine(img, rotation_matrix, (width, height))
plt.imshow(rotated_img)
```

#### Translation

```python
tx = 120
ty = 265
```

```python
translation_matrix = np.float32([[1, 0, tx], [0, 1, ty]])
```

```python
translated_img = cv2.warpAffine(img, translation_matrix, (width, height))
plt.imshow(translated_img)
```

#### Horizontal Reflection

```python
hori_reflect_matrix = np.float32([[1, 0, 0], [0, -1, height], [0, 0, 1]])
```

```python
hori_reflect_img = cv2.warpPerspective(img, hori_reflect_matrix, (width, height))
plt.imshow(hori_reflect_img)
```

#### Vertical Reflection

```python
vert_reflect_matrix = np.float32([[-1, 0, width], [0, 1, 0], [0, 0, 1]])
```

```python
vert_reflect_img = cv2.warpPerspective(img, vert_reflect_matrix, (width, height))
plt.imshow(vert_reflect_img)
```

#### Shearing on X-Axis

```python
shx = 0.75
```

```python
shear_x_matrix = np.float32([[1, shx, 0], [0, 1, 0], [0, 0, 1]])
```

```python
x_sheared_img = cv2.warpPerspective(img, shear_x_matrix, (width, height))
plt.imshow(x_sheared_img)
```

#### Shearing on Y-Axis

```python
shy = 0.37
```

```python
shear_y_matrix = np.float32([[1, 0, 0], [shy, 1, 0], [0, 0, 1]])
```

```python
y_shear_img = cv2.warpPerspective(img, shear_y_matrix, (width, height))
plt.imshow(y_shear_img)
```
